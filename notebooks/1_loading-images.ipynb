## loading image

### with imageio

```python
img = iio.imread(image_path)
```

```python
len_y, len_x, __ = img.shape
```

- pixel `[0, 0]` is at the top left
- pixel `[len_y-1, 0]` is at the bottom left
- pixel `[len_y-1, len_x-1]` is at the bottom right
- pixel `[0, len_x-1]` at at the top right
- pixel color is in RGB format

```python
pos_y, pos_x = 0, 0 # pixel position
rgb = [0, 0, 0] # color in RGB, as an int between 0 and 255

# set color at pixel
img[pos_y, pos_x, :] = rgb
```

One issue is that when saved in jpg-format, saving and loading is not the identity:

```python
# write it
iio.imwrite("test.jpg", img)

# load it again
img2 = iio.imread("test.jpg")

print("Images are the same:", np.isclose(img, img2).all())
```

However, it works when using png format

```python
# write it
iio.imwrite("test.png", img)

# load it again
img2 = iio.imread("test.png")

print("Images are the same:", np.isclose(img, img2).all())
```

### with matplotlib

According to the matplotlib documentation for `matplotlib.image.imread` at https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imread.html, the image loading function only exists for historical reasons. It is suggested to use the pillow library instead. The function is also a bit complicated to use because png files are read with RGB vales between 0 and 1, and if a png image is saved with `matplotlib.imsave` then it's suddenly also getting an alpha value. I'm therefore ignoring `matplotlib` for saving and loading. A nice thing is the plotting function though to see where the pixels are located:

```python
plt.imshow(img)
```
